```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Load dataset
df = pd.read_csv("C:/Users/LENOVO-RKC/OneDrive/Desktop/Super_Store_data.csv", encoding="latin1")

# Convert 'Date' column
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df.set_index('Order Date', inplace=True)

# Aggregate data by month
df_resampled = df.resample('M').sum()  # Change to 'M' for monthly data

# Select column
pollution_col = "Sales"

# Plot rolling statistics
plt.figure(figsize=(12, 6))
plt.plot(df_resampled[pollution_col], label="Original Data", linewidth=2, alpha=0.7)
plt.plot(df_resampled[pollution_col].rolling(window=6).mean(), label="Rolling Mean (6 months)", color='red', linewidth=2)
plt.plot(df_resampled[pollution_col].rolling(window=6).std(), label="Rolling Std Dev (6 months)", color='black', linewidth=2)
plt.legend()
plt.title("Rolling Mean & Standard Deviation (Resampled Monthly)")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
